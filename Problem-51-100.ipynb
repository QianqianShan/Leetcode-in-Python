98 - Validate Binary Search Tree

Given a binary tree, determine if it is a valid binary search tree (BST).

Assume a BST is defined as follows:

* The left subtree of a node contains only nodes with keys less than the node's key.
* The right subtree of a node contains only nodes with keys greater than the node's key.
* Both the left and right subtrees must also be binary search trees.
 

Example:

      2
     / \
    1   3

Input: [2,1,3]

Output: true

```python
# 1. compare the inorder traverse results with sorted values in the tree

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        if not root:
            return True 
        inorder_vals = self.in_order_traverse(root)
        # use set() to make sure that there are no duplicates 
        return inorder_vals == sorted(list(set(inorder_vals)))
    def in_order_traverse(self, root):
        if root is None:
            return []
        return self.in_order_traverse(root.left) + [root.val] + self.in_order_traverse(root.right)
```

```python
# 2. inorder iterate over the tree and compare each node value with its previous node value  

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        if not root:
            return True 
        self.prev = None 
        return self.helper(root)
    
    def helper(self, root):
        if root is None:
            return True 
        # if root is not None 
        
        # iterate over left subtree 
        if not self.helper(root.left):
            return False 
        
        # if left subtree is valid 
        if self.prev is not None and self.prev.val >= root.val:
            return False 
        
        # if both left subtree and root are valid, check right subtree
        
        # update prev node as root 
        self.prev = root 
        return self.helper(root.right)
```

```python
# 3. Recursion 
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        if not root:
            return True 
        return self.valid_tree(root)[0]
    
    def valid_tree(self, root):
        if root is None:
            return True, None, None
        left_ind, left_min, left_max = self.valid_tree(root.left)
        right_ind, right_min, right_max = self.valid_tree(root.right)
        
        # case 1
        if not left_ind or not right_ind:
            return False, None, None 
        # case 2 
        if left_max is not None and root.val <= left_max:
            return False, None, None 
        # case 3
        if right_min is not None and root.val >= right_min:
            return False, None, None 
        
        # if all above three cases are note satisfied, then both left and right subtrees are valid 
        # and the current tree with root as root is also valid 
        min_val = left_min if left_min is not None else root.val 
        max_val = right_max if right_max is not None else root.val 
        return True, min_val, max_val
        
        if left_ind and right_ind and root.val > left_max and root.val < right_min:
            return True, left_min, right_max 
        return False, None, None  
        
```
